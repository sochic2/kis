```
import numpy as np

def MSE(y,t):
  return np.sum((y - t)**2) / 2

def CEE(y,t):
  delta = 1e-7
  return - np.sum(t * np.log(y + delta))
```

```
t = np.array([[0, 0, 1], [0, 1, 0], [1, 0, 0]])
ya = np.array([[0.3, 0.3, 0.4], [0.3, 0.4, 0.3], [0.1, 0.2, 0.7]])
yb = np.array([[0.1, 0.2, 0.7], [0.1, 0.7, 0.2], [0.3, 0.4, 0.3]])
mse_ya = MSE(ya, t)
mse_yb = MSE(yb, t)
cee_ya = CEE(ya, t)
cee_yb = CEE(yb, t)
print(mse_ya, mse_yb)
print(cee_ya, cee_yb)
```

One-hot encoding 분류 문제에서는 정답레이블이 1과 0으로만 이루어져있기 때문에 CEE를 사용하게 되면 1에 해당하는 오류만 산출해냅니다.

정답을 맞추는 것은 1만 제대로 맞추면 되기 때문에 MSE로 0에 해당하는 오류까지 고려하는 것보다 CEE를 이용해서 1의 오류에 집중하는 것이 one-hot encoding 분류문제에 더 좋습니다.
